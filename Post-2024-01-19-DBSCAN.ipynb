```python
import plotly.express as px
from sklearn.datasets import make_blobs

import DBSCANVisualizer as dv
```

```python
# Generate sample data

X, _ = make_blobs(n_samples=200, cluster_std=0.6, random_state=0)
```

```python
fig = px.scatter(x=X[:,0], y=X[:,1])  
fig.update_layout(coloraxis_showscale=False, 
                  width=600, height=400, title='Data')
fig.update_traces(marker=dict(size=8, opacity=0.7))

fig.show()
```

```python
vis = dv.DBSCANVisualizer(X) 

vis.plot_clusters()
```

```python
vis.plot_point_types()
```
