# Isomap Revealed: Transforming Data into Insights

```python
import pandas as pd

import plotly.express as px
```

```python
dfS = pd.read_csv('Swiss_roll.csv')
print(dfS.shape)
dfS.head()
```

```python
# Plot the 3D data points
figS = px.scatter_3d(dfS, x='x', y='y', z='z', color='label',
                     color_discrete_sequence=px.colors.qualitative.D3 )
# Adjusting the view angle    
figS.update_layout(scene_camera=dict(eye=dict(x=1, y=1, z=1.5)),    
                width=600, height=500, 
                title='Swiss Roll Data',
                legend_title='color')
# Decrease the marker size  
figS.update_traces(marker=dict(size=6, opacity=0.6))  
# Show plot
figS.show()
```

```python
from sklearn.manifold import Isomap

modelI = Isomap(n_neighbors=200, n_components=2, eigen_solver='dense')

dfI = pd.DataFrame(modelI.fit_transform(dfS[['x','y','z']]), columns=['x', 'y'])
dfI['label'] = dfS['label']
```

```python
# Plot the 2D data points
figI = px.scatter(dfI, x='x', y='y', color='label', 
                    color_discrete_sequence=px.colors.qualitative.D3)                    
# Adjusting the graph size
figI.update_layout(width=500, height=400,                   
                  legend_title_text='color') 
# Decrease the marker size
figI.update_traces(marker=dict(size=6, opacity=0.7))  
figI.update_yaxes(range=[-0.8, 0.8])
# Show plot
figI.show() 
```
